## 2D Advection
Advection equation vector form:
$\frac{\partial\psi}{\partial t} = -\mathbf{u} \cdot \nabla\psi$

```python
import numpy
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

xmax = 20
ymax = 20

X,Y = numpy.meshgrid(numpy.arange(0,xmax),
                     numpy.arange(0,ymax))

psi = numpy.zeros((xmax,ymax))
psi[:1] = 1.0

u = numpy.zeros((xmax,ymax,2))
for x in range(xmax):
    for y in range(ymax):
        u[x,y,1]=-0.5
```

```python
# Plot the inital conditions
plt.imshow(psi)
cb = plt.colorbar()

plt.figure()
plt.quiver(X, Y, u[:,0], u[:,1])
plt.show()
```
